```python
def countBits(self, n: int) -> List[int]:
        '''s= []
        for i in range(n+1):
            d = bin(i)[2:]
            c = d.count('1')
            s.append(c)
        return s'''

        d = [0]* (n+1)
        s = 1

        for i in range(1,n+1):
            if s * 2 == i:
                s = i
            d[i] = d[i-s] + 1
        return d
```
